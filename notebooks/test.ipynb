```python
# inspect_pt.py
import torch, glob, os
pdir = r'C:\Users\Egor\VsCode project\Scene_graph_localization\res'
files = sorted(glob.glob(os.path.join(pdir, '*.pt')))
print("Found:", len(files))
if len(files)==0:
    raise SystemExit("No .pt files found in " + pdir)

# inspect first few
for f in files:
    d = torch.load(f, weights_only=False)
    print("File:", os.path.basename(f))
    print(" x:", d.get('x').shape if 'x' in d else None)
    print(" edge_index:", d.get('edge_index').shape if 'edge_index' in d else None)
    print(" edge_attr:", d.get('edge_attr').shape if 'edge_attr' in d else None)
    print(" edge_u_class / edge_v_class:", d.get('edge_u_class').shape if 'edge_u_class' in d else None,
          d.get('edge_v_class').shape if 'edge_v_class' in d else None)
    print(" scene_name:", d.get('scene_name'))
    print(" json_path:", d.get('json_path'))
    print("---")
```
